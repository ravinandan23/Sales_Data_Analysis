# **Visualization**

```python
import matplotlib.pyplot as plt

region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='bar')

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()

# Insight
#   The West region generates the highest sales compared to other regions.
```

```python
category_profit = df.groupby('Category')['Profit'].sum()

category_profit.plot(kind='bar')

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.show()

# Insight
#   Technology category produces the highest profit.
#   Some (Furniture) categories may have high sales but low profit.
```

```python
category_quantity = df.groupby('Category')['Quantity'].sum()

category_quantity.plot(kind='bar')

plt.title("Quantity Sold by Category")
plt.xlabel("Category")
plt.ylabel("Quantity")

plt.show()

# Insight
#   Office Supplies usually sell the most quantity.
```

```python
state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)

state_sales.plot(kind='bar')

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")

plt.show()


# Insight
#   These states generate significantly higher revenue than others.
```

```python
plt.scatter(df['Sales'], df['Profit'])

plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()

# Insight
#    Higher sales do not always mean higher profit.
# **This is a very important business insight.**
```


# **Key Insights**

1️ West region has the highest sales performance.

2️ Technology category generates the most profit.

3️ Office supplies sell the highest quantity.

4️ Some states contribute significantly to overall revenue.

5️ High sales do not always guarantee high profit.
